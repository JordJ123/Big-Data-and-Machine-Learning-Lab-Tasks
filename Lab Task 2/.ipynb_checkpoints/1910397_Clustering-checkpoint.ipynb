```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as pyplot
import numpy
```

# Task 1

```python
# IRIS DATA
# 150 Rows =  3 Species x 50 Samples
# 4 Columns = Sepal Length, Sepal Width, 
#             Petal Length, Petal Width

# IRIS LABEL
# Species 0, 1, and 2
```

```python
data = numpy.load('Iris_data.npy')
labels = numpy.load('Iris_labels.npy')
pyplot.figure()
scatter = pyplot.scatter(data[:,0], data[:,1], 
                         c=labels)
elements = scatter.legend_elements()
pyplot.legend(elements[0], ["Setosa", "Versicolor", 
    "Virginica"], title="Species")
pyplot.title("Scatter Plot of Iris Data")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.show()
```

# Task 2

```python
# K_Means has generated 3 clusters
# Clusters fits in the range of the data values
# Predicts what cluster each value is closet to
# by running the K-Means algorithm where the centroid
# are continually moved until it is in the center of
# its related data as much as possible
```

```python
data = numpy.load('Iris_data.npy')
kmeans = KMeans(n_clusters = 3)
scatter = kmeans.fit_predict(data)
clusters = kmeans.cluster_centers_
pyplot.scatter(data[:,0], data[:,1], c=scatter)
pyplot.scatter(clusters[:,0], clusters[:,1], s=150,
    c="red", marker="X")
pyplot.title("K Means")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.show()
```

```python
#init = Method of generating the first centroids
#n_init = Number of times the k-means algorithm is run,
#         each with different starting centroids
data = numpy.load('Iris_data.npy')
kmeans = KMeans(n_clusters = 3, init="random", 
                 n_init=100)
scatter = kmeans.fit_predict(data)
clusters = kmeans.cluster_centers_
pyplot.scatter(data[:,0], data[:,1], c=scatter)
pyplot.scatter(clusters[:,0], clusters[:,1], s=150,
    c="red", marker="X")
pyplot.title("K Means")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.show()
```

# Task 3

```python
# Gaussian Mixture has generated 3 clusters which are 
# fitted in the range of the data values

# Predicts what cluster each value is closet

# Predicts the probabiliy of each value being part of
# each cluster based using a GMM algorithm that uses 
# the probability of a value being part of each cluster,
# the mean, and the variance

# Repeated with updates to cluster probabilties,
# mean and variance until value probabilies don't change

# Cluster that a value has the highest probability
# being a part of is the cluster they are a part of
```

```python
data = numpy.load('Iris_data.npy')
gm = GaussianMixture(n_components=3)
scatter = gm.fit_predict(data)
clusters = gm.means_
pyplot.scatter(data[:,0], data[:,1], c=scatter)
pyplot.scatter(clusters[:,0], clusters[:,1], s=150,
    c="red", marker="X")
pyplot.title("GMM")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.show()
```

```python
data = numpy.load('Iris_data.npy')
gm = GaussianMixture(n_components=3)
gm.fit(data)
scatter = gm.predict_proba(data)

pyplot.scatter(data[:,0], data[:,1], c=scatter[:,0])
pyplot.title("Component 1")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.colorbar()
pyplot.show()

pyplot.scatter(data[:,0], data[:,1], c=scatter[:,1])
pyplot.title("Component 2")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.colorbar()
pyplot.show()

pyplot.scatter(data[:,0], data[:,1], c=scatter[:,2])
pyplot.title("Component 3")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.colorbar()
pyplot.show()
```

```python
#init_params = Determines how the starting values are
# calculated e.g. k-means, random
data = numpy.load('Iris_data.npy')
gm = GaussianMixture(n_components=3, 
    init_params="random")
gm.fit(data)
scatter = gm.predict_proba(data)

pyplot.scatter(data[:,0], data[:,1], c=scatter[:,0])
pyplot.title("Component 1")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.colorbar()
pyplot.show()

pyplot.scatter(data[:,0], data[:,1], c=scatter[:,1])
pyplot.title("Component 2")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.colorbar()
pyplot.show()

pyplot.scatter(data[:,0], data[:,1], c=scatter[:,2])
pyplot.title("Component 3")
pyplot.xlabel("Sepal Length")
pyplot.ylabel("Sepal Width")
pyplot.colorbar()
pyplot.show()
```
